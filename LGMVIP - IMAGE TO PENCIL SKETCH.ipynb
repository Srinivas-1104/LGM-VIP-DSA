## Image to Pencil Sketch with Python

The Basic Steps:

1.    Import the picture
2.    Convert the picture to RBG format
3.    Convert the picture into Grayscale format
4.    Invert the picture into the negative form
5.    Smoothen the inverted picture using blurs
6.    Divide the picture with grayscale and inverted_blur picture 
7.    Obtain the pencil sketch

##  IMPORTS

```python
# From the Python Imaging Library (PIL) import the Image Class
from PIL import Image

# From the IPython.display Module import display module
from IPython.display import display

# Import the cv2 library for Image Processing
import cv2
```

## Reading the Image

```python
image = cv2.imread ("animals.png")
display(Image.fromarray(image))
```

## Convert the image to the RGB format

```python
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
display(Image.fromarray(image_RGB))
```

## Convert the image to Grayscale

```python
grayscale = cv2.cvtColor(image_RGB, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(grayscale))
```

## Invert the Grayscale Image 

```python
invert = 255 - grayscale
display(Image.fromarray(invert))
```

## Smoothen the Image using Blurs

```python
blur = cv2.GaussianBlur(invert, (21, 21), 0)
display(Image.fromarray(blur))
```

## Invert the Blur

```python
invert_blur = 255 - blur
display(Image.fromarray(invert_blur))
```

## Convert to Pencil Sketch

```python
pencil_sketch = cv2.divide(grayscale, invert_blur, scale=256.0)
display(Image.fromarray(pencil_sketch))
```

```python
#  See the path that the image has taken:

display(Image.fromarray(image))
display(Image.fromarray(image_RGB))
display(Image.fromarray(grayscale))
display(Image.fromarray(invert))
display(Image.fromarray(blur))
display(Image.fromarray(invert_blur))
display(Image.fromarray(pencil_sketch))
```
